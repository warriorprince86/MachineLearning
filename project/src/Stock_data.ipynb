```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

from datetime import datetime
from datetime import timedelta
import datetime

import yfinance as yf
from pandas_datareader import data as pdr
```

```python
data = pd.read_csv('data/tsla_2.csv')
dfopen=data[['Date', 'Open']]
dfclose=data[['Date','Adj Close']]
dfopen['Label'] = 'OPEN'
dfopen = dfopen.rename(columns={'Open':'Price'})
dfopen = dfopen[['Date','Label', 'Price']]
dfclose['Label'] = 'CLOSE'
dfclose = dfclose.rename(columns={'Adj Close':'Price'})
dfclose = dfclose[['Date','Label', 'Price']]
```

```python
dfnew=pd.concat([dfopen,dfclose],ignore_index=True)
dfnew
```

```python
dfnew = dfnew.sort_values(['Date', 'Label'],
                          ascending = [True, False])
```
